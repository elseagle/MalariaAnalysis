```python
for c in df1.columns:
    to_remove1 = ['GHOCD', 'GHODP', 'GHOURL', 'PUBCODE', 'PUBDISP', 'PUBURL', 'YEARCD', 'YEARURL','REGURL', 'CNTURL', 'comments']
    for tr in to_remove1:
        if tr in c:
            df1 =df1.drop(tr, axis =1)
df1            
```

```python
# 2013 reported_cases extracted
rep2013= df1[df1['YEAR'] == 2013]
rep2013
```

```python
sum2013 = pd.DataFrame.sum(rep2013['value'])
sum2013
```

```python
for v in rep2013['value']:
    for c in rep2013['REGDISP']:
        percentage = (v/sum2013) * 100
        print(c, '               ',percentage)
```
